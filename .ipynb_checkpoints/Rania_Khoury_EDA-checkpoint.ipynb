```python
def standardize_url(url):
    parsed_url = urlparse(url)
    scheme = parsed_url.scheme
    domain = parsed_url.netloc
    path = parsed_url.path # 
    if path == '/' or path == '':
        return f'{scheme}://{domain}'
    return url
```

```python
sublinks['sublinks'] = sublinks['sublinks'].apply(standardize_url)
```

```python
sublinks.head()
```

```python
def determine_page_type(url):
    path = urlparse(url).path
    if 'blog' in path or 'post' in path:
        return 'Blog'
    elif 'product' in path or 'item' in path or 'collections' in path or 'shop' in path:
        return 'Product'
    elif 'contact' in path:
        return 'Contact'
    elif 'about' in path or 'team' in path:
        return 'About'
    elif 'service' in path or 'project' in path or 'portfolio' in path:
        return 'Service'
    elif 'home' in path or 'index' in path:
        return 'Home'
    elif 'social' in path or 'instagram' in path or 'tiktok' in path or 'facebook' in path:
        return 'Social Media'
    elif 'sign' in path:
        return 'Sign Up Page'
    elif 'workwithus' in path or 'careers' in path or 'jobs' in path:
        return 'Careers Page'
    else:
        return 'Other'
```

```python
sublinks['page_type'] = sublinks['sublinks'].apply(determine_page_type)
```

```python
sublinks['domain'] = sublinks['sublinks'].apply(lambda x: x.split('.')[-1])
```

```python
import re
from collections import Counter
def extract_delimiters(url):
    path = urlparse(url).path
    # Find all non-alphanumeric characters
    delimiters = re.findall(r'[^a-zA-Z0-9]', path)
    return delimiters
```

```python
all_delimiters = [delimiter for sublist in sublinks['sublinks'].apply(extract_delimiters) for delimiter in sublist]
delimiter_counts = Counter(all_delimiters)
print(delimiter_counts.most_common())
```

```python
sublinks_list = sublinks['domain'].tolist()
unique_domains = set(sublinks_list)
average_length_sub_links = sum(len(sub_link) for sub_link in sublinks_list) / len(sublinks_list)
```

```python
sublinks.head()
```

```python
print(f"Average length of sub-links: {average_length_sub_links:.2f}")
```

```python
seed_sublinks = pd.read_csv('results_before_closing.csv')
```

```python
seed_sublinks.head()
```

```python
seed_sublinks.size
```

```python
seed_sublinks.dropna(inplace=True)
```

```python
seed_sublinks['seed_url'] = seed_sublinks['seed_url'].apply(remove_www)
```

```python
seed_sublinks['seed_url'].duplicated().sum()
```

```python
seed_sublinks.drop_duplicates(subset='seed_url', keep='first',inplace=True)
```

```python
seed_sublinks['seed_url'].duplicated().sum()
```

```python
def remove_internal_duplicates(sublinks):
    if pd.isna(sublinks):  
        return sublinks
    sublinks_list = sublinks.split(',')
    unique_sublinks = list(set(sublinks_list))  
    return ','.join(unique_sublinks)
```

```python
seed_sublinks['sub_links'] = seed_sublinks['sub_links'].apply(remove_internal_duplicates)
```

```python
def count_sublinks(sublinks):
    if sublinks == "":
        return 0
    return len(sublinks.split(','))

seed_sublinks['sub_links'] = seed_sublinks['sub_links'].astype(str).fillna('')
seed_sublinks['sublink_count'] = seed_sublinks['sub_links'].apply(count_sublinks)
```

```python
seed_sublinks.head()
```

```python
seed_sublinks = seed_sublinks.sort_values(by=['sublink_count'], ascending=False)
```

```python
print(f'Number of sublinks for each seed URL')
for ind, row in seed_sublinks.iterrows():
    print(f"Seed URL {row['seed_url']}, Sub-links Count: {row['sublink_count']}")
```

```python
seed_sublinks['sublink_count'].describe()
```

```python
def get_first_word(url):
    path = urlparse(url).path
    parts = path.split('/')
    return parts[1] if len(parts) > 1 else 'None'
```

```python
sublinks['first_word'] = sublinks['sublinks'].apply(get_first_word)
```

```python
first_word_counts = sublinks['first_word'].value_counts().reset_index()
first_word_counts.columns = ['first_word', 'count']
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='count', y='first_word', hue='first_word', data=first_word_counts[1:6], palette='deep')
plt.title('Most Common First Words in Sublinks')
plt.xlabel('Count')
plt.ylabel('First Word')
plt.show()
```
